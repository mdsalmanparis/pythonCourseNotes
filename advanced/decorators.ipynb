```python
#Decorators useed to extend the functionality of another function without modifying the base function

#Create a Base Function with Wrapper Function
#Create a Decorator Function

import time 

def time_it(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"Function: {func.__name__} took {end-start:.8f} seconds.")
        if result == None:
            result = "nothing"
        print(f"function returned {result}")
        return result
    return wrapper 

@time_it    
def search(lst, target):
    for num in range(len(lst)):
        if lst[num] == target:
            return "found at index "+str(num)

search([4,5,6,12,78,23,12,9,67],100)
```
