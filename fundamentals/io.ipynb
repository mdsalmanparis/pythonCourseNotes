```python
#The print() Function
print("Hello World!")

#Printing Multiple Values
name = "Alice"
age = 25
print("Name:",name,"Age:",age)

#Formatting Output
print(f"My name is {name} and i'm {age} years old.")

#end & sep 
print("Hello", "World", sep='-', end='!')

#Spacing in outputs
print("\nThis is a new line \n\t -Salman's laptop \n\t -\"this is a quote\" ")
```

```python
#The input() Function
name = input("Enter your name: ")
print(f"Hello, {name}")

#Type Casting 
age = int(input("Enter your age: "))
print("Next year your age will be",age+1)

#Taking Multiple Inputs 
x,y,z = input("Enter 3 values").split()
print(x,y,z)
```
