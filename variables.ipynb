**Rules for Naming Variables**
- Must start with a letter or underscore (_).
- Cannot start with a number.
- Can contain letters, numbers, and underscores.
- Case-sensitive (age and Age are different).
- Avoid Python keywords (class, def, if, etc.).

```python
#Assigning value to a variables using '=' 
radius = 36.89

#Multiple Assignment
y, z = 10,20

#Same value to multiple variables 
a = b = c = 100
```

```python
#Type Checking 
print(type(radius))

#Deleting Variables
del radius
```

```python
#Dynamic Typing
x = 10       # int
x = "hello"  # str

print(x)
```

```python
x = "global"

def func():
    x = "local"
    print(x)

func()   # local
print(x) # global
```

##### Memory References of an Variables.
- Variables are references to objects in the memory 
- if the values are same it will point to the same objects 

```python
a = 100
b = 100

print(id(a))
print(id(b))
```

```python
list1 = [1,2,3]
list2 = list1
list2.append(4)
print(list1)
```
